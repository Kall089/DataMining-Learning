## 🔍 1. Apa itu Ukuran Tendensi Sentral?

**Ukuran tendensi sentral** (measures of central tendency) adalah angka yang merepresentasikan **"pusat"** atau **"nilai tipikal"** dari sebuah dataset.

### Analogi Sederhana:
Bayangkan kamu diminta mendeskripsikan tinggi badan teman-teman sekelas dalam **satu angka**. Angka itu adalah ukuran tendensi sentral.

### Tiga Ukuran Utama:

| Ukuran | Definisi | Kapan Pakai | Kelemahan |
|--------|----------|-------------|------------|
| **Mean** | Rata-rata aritmatika | Data simetris, tanpa outlier | Sensitif terhadap outlier |
| **Median** | Nilai tengah saat data diurutkan | Data skewed, ada outlier | Membuang informasi distribusi |
| **Mode** | Nilai yang paling sering muncul | Data kategorik, multimodal | Bisa tidak ada, atau > 1 |

### Pertanyaan Kunci:
> **"Jika kamu harus meringkas seluruh dataset ke satu angka, angka mana yang paling mewakili?"**

Jawabannya TERGANTUNG pada bentuk distribusi data!

## 🔍 2. Mean (Rata-Rata Aritmatika)

### Definisi:
Mean adalah **jumlah seluruh nilai dibagi banyaknya data**.

### 🧮 Rumus:
$$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i = \frac{x_1 + x_2 + ... + x_n}{n}$$

### Intuisi:
- Mean adalah **titik keseimbangan** (balance point) dari data
- Jika data adalah batu-batu di papan jungkit, mean adalah titik di mana papan seimbang
- Setiap data punya **pengaruh yang sama** terhadap mean

### Sifat Mean:
1. **Menggunakan semua data** — tidak ada yang terbuang
2. **Sensitif terhadap outlier** — 1 nilai ekstrem bisa menggeser mean jauh
3. **Unik** — setiap dataset punya tepat 1 mean
4. **Jumlah deviasi = 0** — $\sum(x_i - \bar{x}) = 0$ selalu

### Kapan Pakai Mean:
- Data berdistribusi **simetris** (tidak skewed)
- Tidak ada **outlier** yang ekstrem
- Data bertipe **interval/rasio** (numerik)

````python
# Demo: Menghitung Mean
# Dataset: Nilai ujian 10 mahasiswa
nilai_ujian = np.array([75, 80, 82, 78, 85, 79, 81, 77, 83, 80])

# Hitung manual
mean_manual = sum(nilai_ujian) / len(nilai_ujian)

# Hitung dengan NumPy dan Pandas
mean_numpy = np.mean(nilai_ujian)
mean_pandas = pd.Series(nilai_ujian).mean()

print('Dataset Nilai Ujian:')
print(f'  Data: {nilai_ujian}')
print(f'  Jumlah data (n): {len(nilai_ujian)}')
print(f'  Sum: {sum(nilai_ujian)}')
print()
print(f'Mean (manual)  : {sum(nilai_ujian)} / {len(nilai_ujian)} = {mean_manual:.2f}')
print(f'Mean (numpy)   : {mean_numpy:.2f}')
print(f'Mean (pandas)  : {mean_pandas:.2f}')
print()
print('Interpretasi: Rata-rata nilai ujian kelas ini adalah {:.1f}'.format(mean_numpy))
print('  → Nilai "tipikal" mahasiswa di kelas ini sekitar 80')
````

### ⚠️ Masalah Mean: Sensitif Terhadap Outlier

Sekarang bayangkan ada **satu mahasiswa yang nilainya sangat rendah** (mungkin tidak masuk ujian). Lihat apa yang terjadi pada mean:

````python
# Demo: Mean MENIPU karena outlier
nilai_normal = np.array([75, 80, 82, 78, 85, 79, 81, 77, 83, 80])
nilai_dengan_outlier = np.array([75, 80, 82, 78, 85, 79, 81, 77, 83, 15])  # 15 = outlier!

mean_normal = np.mean(nilai_normal)
mean_outlier = np.mean(nilai_dengan_outlier)

print('=== TANPA Outlier ===')
print(f'  Data: {nilai_normal}')
print(f'  Mean: {mean_normal:.2f}')
print()
print('=== DENGAN Outlier (80 → 15) ===')
print(f'  Data: {nilai_dengan_outlier}')
print(f'  Mean: {mean_outlier:.2f}')
print()
print(f'⚠️ Perubahan mean: {mean_normal:.2f} → {mean_outlier:.2f} (turun {mean_normal - mean_outlier:.2f} poin!)')
print(f'  Hanya SATU outlier menggeser mean {((mean_normal - mean_outlier)/mean_normal*100):.1f}%')
print()
print('Apakah 73.5 mewakili "nilai tipikal" kelas ini? TIDAK!')
print('9 dari 10 mahasiswa nilainya > 75, tapi mean bilang "rata-rata 73.5"')
````

````python
# Visualisasi: Mean tertarik oleh outlier
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot TANPA outlier
axes[0].hist(nilai_normal, bins=8, color='#2196F3', edgecolor='black', alpha=0.7)
axes[0].axvline(np.mean(nilai_normal), color='red', linestyle='--', linewidth=2, label=f'Mean = {np.mean(nilai_normal):.1f}')
axes[0].axvline(np.median(nilai_normal), color='green', linestyle='-', linewidth=2, label=f'Median = {np.median(nilai_normal):.1f}')
axes[0].set_title('Tanpa Outlier: Mean ≈ Median', fontweight='bold', fontsize=13)
axes[0].set_xlabel('Nilai Ujian')
axes[0].set_ylabel('Frekuensi')
axes[0].legend(fontsize=11)
axes[0].set_xlim(10, 90)

# Plot DENGAN outlier
axes[1].hist(nilai_dengan_outlier, bins=8, color='#FF9800', edgecolor='black', alpha=0.7)
axes[1].axvline(np.mean(nilai_dengan_outlier), color='red', linestyle='--', linewidth=2, label=f'Mean = {np.mean(nilai_dengan_outlier):.1f}')
axes[1].axvline(np.median(nilai_dengan_outlier), color='green', linestyle='-', linewidth=2, label=f'Median = {np.median(nilai_dengan_outlier):.1f}')
axes[1].set_title('Dengan Outlier: Mean TERTARIK ke kiri!', fontweight='bold', fontsize=13)
axes[1].set_xlabel('Nilai Ujian')
axes[1].set_ylabel('Frekuensi')
axes[1].legend(fontsize=11)
axes[1].set_xlim(10, 90)

plt.suptitle('Efek Outlier pada Mean vs Median', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

## 🔍 3. Median (Nilai Tengah)

### Definisi:
Median adalah **nilai yang berada tepat di tengah** ketika data sudah diurutkan dari terkecil ke terbesar.

### 🧮 Cara Hitung:
1. **Urutkan** data dari kecil ke besar
2. Jika n **ganjil**: median = nilai di posisi $(n+1)/2$
3. Jika n **genap**: median = rata-rata 2 nilai di tengah

$$\text{Median} = \begin{cases} x_{(n+1)/2} & \text{jika n ganjil} \\ \frac{x_{n/2} + x_{n/2+1}}{2} & \text{jika n genap} \end{cases}$$

### Intuisi:
- Median **membagi data menjadi 2 bagian yang sama besar** (50% di bawah, 50% di atas)
- Median **tidak peduli** seberapa ekstrem outlier — hanya posisi yang penting
- Inilah mengapa median disebut **resistant statistic** (tahan terhadap outlier)

### Sifat Median:
1. **Resistant** — tidak terpengaruh outlier
2. **Hanya bergantung pada posisi** — bukan nilai
3. **Kurang informatif** — membuang informasi dari data ekstrem
4. **Lebih stabil** — untuk sampel kecil dari populasi skewed

### Kapan Pakai Median:
- Data **skewed** (miring ke kiri/kanan)
- Ada **outlier** yang bukan error tapi data valid
- Ingin ukuran yang **robust** terhadap extreme values
- Contoh: pendapatan rumah tangga, harga rumah, waktu respons server

````python
# Demo: Menghitung Median

# Kasus n GANJIL
data_ganjil = np.array([3, 7, 1, 9, 5])
data_ganjil_sorted = np.sort(data_ganjil)
n = len(data_ganjil)
pos = (n + 1) // 2  # posisi median (1-indexed)

print('=== Kasus n Ganjil ===')
print(f'  Data asli    : {data_ganjil}')
print(f'  Data terurut : {data_ganjil_sorted}')
print(f'  n = {n}, posisi tengah = ({n}+1)/2 = {(n+1)/2:.0f}')
print(f'  Median = {data_ganjil_sorted[pos-1]} (nilai di posisi ke-{pos})')
print(f'  NumPy: np.median() = {np.median(data_ganjil)}')
print()

# Kasus n GENAP
data_genap = np.array([3, 7, 1, 9, 5, 8])
data_genap_sorted = np.sort(data_genap)
n = len(data_genap)
pos1, pos2 = n // 2, n // 2 + 1  # 1-indexed

print('=== Kasus n Genap ===')
print(f'  Data asli    : {data_genap}')
print(f'  Data terurut : {data_genap_sorted}')
print(f'  n = {n}, posisi tengah = {n//2} dan {n//2 + 1}')
print(f'  Median = ({data_genap_sorted[pos1-1]} + {data_genap_sorted[pos2-1]}) / 2 = {(data_genap_sorted[pos1-1] + data_genap_sorted[pos2-1])/2}')
print(f'  NumPy: np.median() = {np.median(data_genap)}')
````

````python
# Demo: Median RESISTANT terhadap outlier
# Real-world case: Gaji karyawan + CEO

gaji_karyawan = np.array([5, 6, 5.5, 7, 6.5, 5, 6, 7.5, 5.5, 6])  # dalam juta Rupiah
gaji_dengan_ceo = np.append(gaji_karyawan, 150)  # CEO gaji 150 juta

print('=== Gaji Perusahaan (dalam juta Rupiah) ===')
print(f'  10 Karyawan: {gaji_karyawan}')
print(f'  + CEO      : {gaji_dengan_ceo[-1]} juta')
print()
print('--- Tanpa CEO ---')
print(f'  Mean   : Rp {np.mean(gaji_karyawan):.2f} juta')
print(f'  Median : Rp {np.median(gaji_karyawan):.2f} juta')
print()
print('--- Dengan CEO ---')
print(f'  Mean   : Rp {np.mean(gaji_dengan_ceo):.2f} juta  ← MENIPU!')
print(f'  Median : Rp {np.median(gaji_dengan_ceo):.2f} juta  ← Masih realistis')
print()
print('💡 Insight:')
print(f'  Mean bilang "rata-rata gaji {np.mean(gaji_dengan_ceo):.1f} juta"')
print(f'  Tapi 10 dari 11 orang gajinya < {np.mean(gaji_dengan_ceo):.0f} juta!')
print(f'  Median ({np.median(gaji_dengan_ceo):.1f} juta) lebih mewakili "gaji tipikal"')
````

## 🔍 4. Mode (Nilai Paling Sering Muncul)

### Definisi:
Mode adalah **nilai yang paling sering muncul** (frekuensi tertinggi) dalam dataset.

### Sifat Mode:
1. **Bisa tidak ada** — jika semua nilai muncul dengan frekuensi sama
2. **Bisa lebih dari satu** — bimodal (2 mode), multimodal (> 2 mode)
3. **Berlaku untuk semua tipe data** — termasuk kategorik!
4. **Tidak terpengaruh outlier** — karena hanya menghitung frekuensi

### Kapan Pakai Mode:
- Data **kategorik** (nominal/ordinal) — mean dan median tidak bisa dihitung!
- Ingin tahu nilai yang **paling umum/populer**
- Deteksi **distribusi multimodal** (ada > 1 puncak)
- Contoh: ukuran baju terlaris, metode bayar paling populer, jam sibuk

### Kelemahan Mode:
- Bisa **tidak informatif** untuk data kontinu (setiap nilai mungkin unik)
- Tidak mempertimbangkan **nilai** data, hanya frekuensi
- **Tidak stabil** — perubahan kecil bisa mengubah mode

````python
# Demo: Mode

# Mode data numerik
data_numerik = np.array([1, 2, 2, 3, 3, 3, 4, 4, 5])
mode_result = stats.mode(data_numerik, keepdims=True)

print('=== Mode Data Numerik ===')
print(f'  Data: {data_numerik}')
print(f'  Mode (SciPy): {mode_result.mode[0]} (muncul {mode_result.count[0]}x)')
print(f'  Mode (Pandas): {pd.Series(data_numerik).mode().values}')
print()

# Mode data kategorik — INI KASUS UTAMA MODE
metode_bayar = ['qris', 'transfer', 'qris', 'cod', 'qris', 'transfer', 'qris', 'cod', 'transfer', 'qris']
series_bayar = pd.Series(metode_bayar)

print('=== Mode Data Kategorik ===')
print(f'  Data: {metode_bayar}')
print(f'  Mode: {series_bayar.mode()[0]}')
print(f'  Value counts:')
print(series_bayar.value_counts().to_string())
print()
print('💡 Mean dan median TIDAK BISA dihitung untuk data kategorik!')
print('   Mode adalah satu-satunya ukuran central tendency yang valid.')
print()

# Kasus BIMODAL
data_bimodal = np.array([1, 1, 1, 2, 3, 5, 5, 5, 6, 7])
print('=== Kasus Bimodal ===')
print(f'  Data: {data_bimodal}')
print(f'  Modes: {pd.Series(data_bimodal).mode().values}  ← Ada 2 mode!')
print('  Artinya distribusi data punya DUA puncak (bimodal)')
````

````python
# Visualisasi: Distribusi Unimodal vs Bimodal
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Unimodal
np.random.seed(42)
data_uni = np.random.normal(50, 10, 500)
axes[0].hist(data_uni, bins=25, color='#4CAF50', edgecolor='black', alpha=0.7)
axes[0].axvline(np.mean(data_uni), color='red', linestyle='--', lw=2, label=f'Mean = {np.mean(data_uni):.1f}')
axes[0].set_title('Unimodal: Satu Puncak', fontweight='bold', fontsize=13)
axes[0].set_xlabel('Nilai')
axes[0].set_ylabel('Frekuensi')
axes[0].legend()

# Bimodal
data_bi = np.concatenate([np.random.normal(30, 5, 250), np.random.normal(70, 5, 250)])
axes[1].hist(data_bi, bins=30, color='#FF5722', edgecolor='black', alpha=0.7)
axes[1].axvline(np.mean(data_bi), color='red', linestyle='--', lw=2, label=f'Mean = {np.mean(data_bi):.1f}')
axes[1].axvline(30, color='blue', linestyle=':', lw=2, label='Mode 1 ≈ 30')
axes[1].axvline(70, color='blue', linestyle=':', lw=2, label='Mode 2 ≈ 70')
axes[1].set_title('Bimodal: Dua Puncak — Mean di "lembah"!', fontweight='bold', fontsize=13)
axes[1].set_xlabel('Nilai')
axes[1].set_ylabel('Frekuensi')
axes[1].legend()

plt.suptitle('Mode Mendeteksi Bentuk Distribusi yang Mean Tidak Bisa', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print('💡 Pada distribusi bimodal, mean jatuh di LEMBAH (antara 2 puncak)')
print('   Tidak ada satu pun data point yang dekat dengan mean!')
print('   Dalam kasus ini, mode JAUH lebih informatif daripada mean.')
````

## 🔍 5. Resistant Statistics — Mengapa Ini Penting

### Definisi:
**Resistant statistic** (atau robust statistic) adalah ukuran statistik yang **tidak berubah drastis** jika ada sedikit data ekstrem (outlier).

### Perbandingan:

| Statistik | Resistant? | Mengapa |
|-----------|-----------|----------|
| **Mean** | ❌ TIDAK | Satu outlier bisa menggeser mean jauh |
| **Median** | ✅ YA | Hanya bergantung pada posisi, bukan nilai |
| **Mode** | ✅ YA | Hanya menghitung frekuensi |
| **Trimmed Mean** | ✅ YA (sebagian) | Menghilangkan extreme values sebelum hitung mean |
| **Range** | ❌ TIDAK | Range = max - min, langsung berubah jika outlier |
| **IQR** | ✅ YA | Q3 - Q1, tidak terpengaruh outlier |

### Real-World Importance:
Di dunia nyata, data SELALU kotor. Ada typo, ada outlier, ada error sensor. Ukuran yang resistant memberikan gambaran yang lebih **jujur** tentang data.

### Breakdown Point:
- **Breakdown point** = proporsi data yang bisa "rusak" sebelum statistik menjadi tidak informatif
- Mean: breakdown point = 0% (satu outlier bisa merusak)
- Median: breakdown point = 50% (setengah data harus rusak untuk merusak median)

````python
# Demo: Resistance Test
# Kita mulai dari data normal, lalu tambah outlier satu per satu

np.random.seed(42)
data_asli = np.random.normal(100, 10, 50)  # 50 data, mean ≈ 100

# Tambah outlier progresif
outliers = [500, 1000, 5000, 10000]  # Makin ekstrem
results = []

for outlier in outliers:
    data_test = np.append(data_asli, outlier)
    results.append({
        'outlier': outlier,
        'mean': np.mean(data_test),
        'median': np.median(data_test),
        'mean_change': np.mean(data_test) - np.mean(data_asli),
        'median_change': np.median(data_test) - np.median(data_asli),
    })

df_resistant = pd.DataFrame(results)
print(f'Data asli: 50 angka dari Normal(100, 10)')
print(f'Mean asli : {np.mean(data_asli):.2f}')
print(f'Median asli: {np.median(data_asli):.2f}')
print()
print('Setelah menambahkan SATU outlier:')
print('=' * 65)
print(f'{"Outlier":>10} | {"Mean":>10} | {"Median":>10} | {"Δ Mean":>10} | {"Δ Median":>10}')
print('-' * 65)
for _, row in df_resistant.iterrows():
    print(f'{row["outlier"]:>10.0f} | {row["mean"]:>10.2f} | {row["median"]:>10.2f} | {row["mean_change"]:>+10.2f} | {row["median_change"]:>+10.2f}')
print()
print('💡 Median hampir TIDAK BERUBAH meskipun outlier makin ekstrem!')
print('   Mean terus naik proporsional dengan outlier.')
````

````python
# Visualisasi: Sensitivity Test
fig, ax = plt.subplots(figsize=(10, 6))

outlier_values = [0] + outliers
mean_values = [np.mean(data_asli)] + [r['mean'] for r in results]
median_values = [np.median(data_asli)] + [r['median'] for r in results]

ax.plot(range(len(outlier_values)), mean_values, 'ro-', markersize=10, linewidth=2, label='Mean')
ax.plot(range(len(outlier_values)), median_values, 'gs-', markersize=10, linewidth=2, label='Median')

ax.set_xticks(range(len(outlier_values)))
ax.set_xticklabels(['No outlier'] + [f'+{o}' for o in outliers])
ax.set_xlabel('Outlier yang Ditambahkan', fontsize=12)
ax.set_ylabel('Nilai Statistik', fontsize=12)
ax.set_title('Resistance Test: Mean vs Median', fontweight='bold', fontsize=14)
ax.legend(fontsize=12)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 🔍 6. Trimmed Mean — Kompromi Cerdas

### Definisi:
**Trimmed mean** (rata-rata terpangkas) menghitung mean setelah **membuang persentase tertentu** dari data di kedua ujung (terkecil dan terbesar).

### 🧮 Cara Kerja:
1. Urutkan data
2. Buang `p%` data dari ujung bawah DAN ujung atas
3. Hitung mean dari data yang tersisa

$$\text{Trimmed Mean}_{p} = \text{mean}(x_{[np]+1}, ..., x_{n-[np]})$$

### Contoh:
- **10% trimmed mean**: buang 10% terkecil + 10% terbesar, hitung mean dari 80% sisanya
- **25% trimmed mean** = **Interquartile Mean** (mean dari data antara Q1 dan Q3)
- **50% trimmed mean** = **Median!** (buang semua kecuali 1-2 nilai tengah)

### Kapan Pakai:
- Ingin **robustness** median tapi **tidak ingin membuang informasi** sebanyak median
- Data punya **sedikit outlier** di ujung, tapi bagian tengah normal
- Kompetisi olahraga (figure skating, diving) — skor tertinggi dan terendah dibuang!

### Keuntungan:
- Lebih robust dari mean biasa
- Lebih informatif dari median (menggunakan lebih banyak data)
- Parameter `p` bisa disesuaikan dengan kebutuhan

````python
# Demo: Trimmed Mean
from scipy.stats import trim_mean

# Simulasi: Skor juri kompetisi menari
# 1 juri memberi skor terlalu tinggi, 1 juri terlalu rendah
skor_juri = np.array([3.0, 7.5, 8.0, 8.2, 8.5, 8.3, 8.1, 8.4, 8.0, 9.8])
skor_sorted = np.sort(skor_juri)

print('=== Skor Juri Kompetisi Menari ===')
print(f'  Skor (urut): {skor_sorted}')
print(f'  Outlier rendah: 3.0 (mungkin bias/error)')
print(f'  Outlier tinggi: 9.8 (mungkin bias/error)')
print()
print(f'  Mean biasa      : {np.mean(skor_juri):.3f}')
print(f'  Median           : {np.median(skor_juri):.3f}')
print(f'  Trimmed mean 10%: {trim_mean(skor_juri, 0.1):.3f}  ← Buang 1 dari tiap ujung')
print(f'  Trimmed mean 20%: {trim_mean(skor_juri, 0.2):.3f}  ← Buang 2 dari tiap ujung')
print()
print('💡 Trimmed mean 10% ≈ menghitung mean setelah buang skor bias')
print('   Ini praktik nyata di Olympic figure skating & diving!')

# Manual calculation untuk verifikasi
n_trim = int(0.1 * len(skor_juri))  # 10% dari 10 = 1
data_trimmed = skor_sorted[n_trim:-n_trim]  # buang 1 dari tiap ujung
print(f'\n  Verifikasi manual: mean({data_trimmed}) = {np.mean(data_trimmed):.3f} ✓')
````

## 🔍 7. Weighted Mean — Rata-Rata Berbobot

### Definisi:
**Weighted mean** menghitung rata-rata di mana setiap data punya **bobot (weight) berbeda** — data yang lebih "penting" berkontribusi lebih besar.

### 🧮 Rumus:
$$\bar{x}_w = \frac{\sum_{i=1}^{n} w_i \cdot x_i}{\sum_{i=1}^{n} w_i}$$

### Contoh Real-World:
- **IPK**: Nilai mata kuliah × SKS (MK 4 SKS lebih berpengaruh dari 2 SKS)
- **Harga rata-rata saham** (weighted by volume): Saham yang banyak diperdagangkan lebih berpengaruh
- **Customer satisfaction** (weighted by spending): Customer yang belanja banyak lebih penting untuk bisnis
- **Indeks harga** (CPI): Barang pokok berbobot lebih besar dari barang mewah

### Perbedaan dengan Mean Biasa:
- Mean biasa: semua data punya bobot sama ($w_i = 1$ untuk semua $i$)
- Weighted mean: bobot bisa berbeda-beda

````python
# Demo: Weighted Mean — Kasus IPK

# Data mata kuliah semester ini
mk_data = pd.DataFrame({
    'mata_kuliah': ['Data Mining', 'Statistika', 'Algoritma', 'Bahasa Inggris', 'Olahraga'],
    'sks':         [4, 3, 4, 2, 1],
    'nilai':       [3.7, 3.5, 3.8, 4.0, 3.0],  # Skala 4.0
})

print('=== Perhitungan IPK (Weighted Mean) ===')
display(mk_data)

# Mean biasa (SALAH untuk IPK!)
mean_biasa = mk_data['nilai'].mean()

# Weighted mean (BENAR)
weighted_mean = np.average(mk_data['nilai'], weights=mk_data['sks'])

# Verifikasi manual
mk_data['kontribusi'] = mk_data['sks'] * mk_data['nilai']
ipk_manual = mk_data['kontribusi'].sum() / mk_data['sks'].sum()

print(f'\nMean biasa (tanpa bobot)  : {mean_biasa:.4f}  ← SALAH! Olahraga = Data Mining?')
print(f'Weighted mean (IPK benar): {weighted_mean:.4f}  ← BENAR!')
print(f'Verifikasi manual        : {ipk_manual:.4f}')
print()
print('Detail perhitungan:')
print(f'  Σ(SKS × Nilai) = {mk_data["kontribusi"].sum():.1f}')
print(f'  Σ(SKS) = {mk_data["sks"].sum()}')
print(f'  IPK = {mk_data["kontribusi"].sum():.1f} / {mk_data["sks"].sum()} = {ipk_manual:.4f}')
print()
print('💡 Data Mining (4 SKS, nilai 3.7) berpengaruh 4x lipat dibanding')
print('   Olahraga (1 SKS, nilai 3.0) terhadap IPK!')
````

## 📊 8. Hubungan Mean, Median, dan Skewness

### Aturan Penting:

| Kondisi | Bentuk Distribusi | Penjelasan |
|---------|-------------------|------------|
| Mean ≈ Median | **Simetris** | Data seimbang, mean dan median hampir sama |
| Mean > Median | **Right-skewed** (positif) | Ekor panjang ke kanan, ada nilai besar yang menarik mean |
| Mean < Median | **Left-skewed** (negatif) | Ekor panjang ke kiri, ada nilai kecil yang menarik mean |

### Intuisi:
- Mean **ditarik ke arah ekor** (tail) distribusi
- Median **tetap di tengah** tidak peduli bentuk ekor
- Gap antara mean dan median = **indikator skewness**

### Real-World Patterns:
- **Right-skewed**: pendapatan, harga rumah, followers social media, waktu tunggu
- **Left-skewed**: nilai ujian (kelas pintar), usia kematian di negara maju
- **Simetris**: tinggi badan, berat badan, IQ, suhu harian

````python
# Visualisasi: Tiga Bentuk Distribusi
np.random.seed(42)

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Left-skewed (negatif)
data_left = np.random.beta(8, 2, 1000) * 100
axes[0].hist(data_left, bins=30, color='#9C27B0', edgecolor='black', alpha=0.7, density=True)
axes[0].axvline(np.mean(data_left), color='red', linestyle='--', lw=2.5, label=f'Mean = {np.mean(data_left):.1f}')
axes[0].axvline(np.median(data_left), color='green', linestyle='-', lw=2.5, label=f'Median = {np.median(data_left):.1f}')
axes[0].set_title('Left-Skewed (Mean < Median)\nContoh: Nilai ujian kelas pintar', fontweight='bold', fontsize=11)
axes[0].legend(fontsize=10)
axes[0].set_xlabel('Nilai')

# Simetris
data_sym = np.random.normal(50, 10, 1000)
axes[1].hist(data_sym, bins=30, color='#2196F3', edgecolor='black', alpha=0.7, density=True)
axes[1].axvline(np.mean(data_sym), color='red', linestyle='--', lw=2.5, label=f'Mean = {np.mean(data_sym):.1f}')
axes[1].axvline(np.median(data_sym), color='green', linestyle='-', lw=2.5, label=f'Median = {np.median(data_sym):.1f}')
axes[1].set_title('Simetris (Mean ≈ Median)\nContoh: Tinggi badan', fontweight='bold', fontsize=11)
axes[1].legend(fontsize=10)
axes[1].set_xlabel('Nilai')

# Right-skewed (positif)
data_right = np.random.exponential(20, 1000) + 10
axes[2].hist(data_right, bins=30, color='#FF5722', edgecolor='black', alpha=0.7, density=True)
axes[2].axvline(np.mean(data_right), color='red', linestyle='--', lw=2.5, label=f'Mean = {np.mean(data_right):.1f}')
axes[2].axvline(np.median(data_right), color='green', linestyle='-', lw=2.5, label=f'Median = {np.median(data_right):.1f}')
axes[2].set_title('Right-Skewed (Mean > Median)\nContoh: Pendapatan, harga rumah', fontweight='bold', fontsize=11)
axes[2].legend(fontsize=10)
axes[2].set_xlabel('Nilai')

plt.suptitle('Hubungan Mean, Median, dan Skewness', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print('Skewness values:')
print(f'  Left-skewed : {pd.Series(data_left).skew():.4f} (negatif)')
print(f'  Simetris    : {pd.Series(data_sym).skew():.4f} (mendekati 0)')
print(f'  Right-skewed: {pd.Series(data_right).skew():.4f} (positif)')
````

## 🎯 9. Kapan Pakai Apa? — Decision Guide

### Flowchart Memilih Ukuran Central Tendency:

```
Apakah data KATEGORIK?
├── YA → Pakai MODE
└── TIDAK (numerik) →
    Apakah distribusi SIMETRIS & tanpa outlier?
    ├── YA → Pakai MEAN (paling informatif)
    └── TIDAK →
        Apakah ada OUTLIER atau data SKEWED?
        ├── YA → Pakai MEDIAN
        │        (atau TRIMMED MEAN jika ingin lebih informatif)
        └── Apakah data BIMODAL/MULTIMODAL?
            └── YA → Pakai MODE (atau analisis per-cluster)
```

### Quick Decision Table:

| Situasi | Pilihan Terbaik | Alasan |
|---------|-----------------|--------|
| Gaji karyawan perusahaan | **Median** | Skewed right karena gaji eksekutif |
| Tinggi badan populasi | **Mean** | Distribusi simetris (normal) |
| Ukuran baju terlaris | **Mode** | Data kategorik (S, M, L, XL) |
| Harga rumah di kota | **Median** | Skewed right (rumah mewah = outlier) |
| Suhu harian | **Mean** | Distribusi mendekati normal |
| Skor juri kompetisi | **Trimmed mean** | Menghilangkan bias juri |
| IPK mahasiswa | **Weighted mean** | Bobot SKS berbeda-beda |
| Rating produk (1-5) | **Median/Mode** | Distribusi biasanya skewed |
| Waktu response API | **Median (P50)** | Right-skewed, outlier banyak |

````python
# Demo Real-World Case: Pendapatan Rumah Tangga Indonesia (Simulasi)
np.random.seed(42)

# Simulasi distribusi pendapatan (log-normal — tipikal distribusi pendapatan)
n_rumah_tangga = 10000
pendapatan = np.random.lognormal(mean=np.log(5), sigma=0.8, size=n_rumah_tangga)  # dalam juta

print('=== Simulasi Pendapatan Rumah Tangga (n=10,000) ===')
print(f'  Mean       : Rp {np.mean(pendapatan):.2f} juta/bulan')
print(f'  Median     : Rp {np.median(pendapatan):.2f} juta/bulan')
print(f'  Mode (approx): Rp {pd.Series(pendapatan).round(0).mode()[0]:.0f} juta/bulan')
print(f'  Trimmed 10%: Rp {trim_mean(pendapatan, 0.1):.2f} juta/bulan')
print(f'  Skewness   : {pd.Series(pendapatan).skew():.4f} (right-skewed!)')
print()
print(f'  Min: Rp {np.min(pendapatan):.2f} juta')
print(f'  Max: Rp {np.max(pendapatan):.2f} juta')
print(f'  % di bawah mean: {(pendapatan < np.mean(pendapatan)).mean()*100:.1f}%')
print()
print('💡 INSIGHT:')
print(f'  {(pendapatan < np.mean(pendapatan)).mean()*100:.0f}% rumah tangga punya pendapatan DI BAWAH "rata-rata"!')
print('  Jika pemerintah pakai MEAN untuk kebijakan, mayoritas akan dirugikan.')
print('  MEDIAN jauh lebih mewakili "pendapatan tipikal" rakyat.')
````

````python
# Visualisasi distribusi pendapatan
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Histogram
axes[0].hist(pendapatan, bins=50, color='#4CAF50', edgecolor='black', alpha=0.7, density=True)
axes[0].axvline(np.mean(pendapatan), color='red', linestyle='--', lw=2.5, label=f'Mean = {np.mean(pendapatan):.1f} juta')
axes[0].axvline(np.median(pendapatan), color='blue', linestyle='-', lw=2.5, label=f'Median = {np.median(pendapatan):.1f} juta')
axes[0].axvline(trim_mean(pendapatan, 0.1), color='orange', linestyle=':', lw=2.5, label=f'Trim 10% = {trim_mean(pendapatan, 0.1):.1f} juta')
axes[0].set_title('Distribusi Pendapatan Rumah Tangga', fontweight='bold', fontsize=12)
axes[0].set_xlabel('Pendapatan (juta Rp/bulan)')
axes[0].set_ylabel('Density')
axes[0].legend(fontsize=10)
axes[0].set_xlim(0, 30)

# Boxplot
axes[1].boxplot(pendapatan, vert=True, widths=0.5)
axes[1].axhline(np.mean(pendapatan), color='red', linestyle='--', lw=2, label=f'Mean = {np.mean(pendapatan):.1f}')
axes[1].axhline(np.median(pendapatan), color='blue', linestyle='-', lw=2, label=f'Median = {np.median(pendapatan):.1f}')
axes[1].set_title('Boxplot Pendapatan', fontweight='bold', fontsize=12)
axes[1].set_ylabel('Pendapatan (juta Rp/bulan)')
axes[1].legend(fontsize=10)

plt.suptitle('Mengapa Median Lebih Tepat untuk Data Pendapatan?', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

## ⚠️ 10. Common Mistakes & Pitfalls

### Mistake 1: "Mean selalu yang terbaik"
❌ **SALAH**: Mean adalah yang paling sering diajarkan, tapi BUKAN selalu yang terbaik.

### Mistake 2: "Median untuk semua"
❌ **SALAH**: Median membuang informasi. Jika data simetris, mean lebih efisien.

### Mistake 3: "Mean dan Median pasti berbeda untuk skewed data"
⚠️ **HATI-HATI**: Perbedaan mean-median memang indikator skewness, tapi bukan satu-satunya. Ada distribusi yang skewed tapi mean ≈ median.

### Mistake 4: "Outlier selalu harus dibuang"
❌ **SALAH**: Outlier bisa jadi data VALID (gaji CEO, harga rumah mewah). Jangan buang tanpa justifikasi!

### Mistake 5: "Mode tidak berguna untuk data numerik"
❌ **SALAH**: Mode mendeteksi distribusi **bimodal** yang mean dan median tidak bisa tangkap.

### Golden Rule:
> **"Selalu PLOT data sebelum memilih statistik deskriptif. Mata manusia mendeteksi anomali lebih cepat dari angka."**

````python
# Demo: Kasus di mana Mean, Median, Mode SEMUA menipu
# Dataset bimodal: campuran 2 populasi berbeda

np.random.seed(42)

# Karyawan part-time (gaji rendah) + full-time (gaji tinggi)
gaji_parttime = np.random.normal(3, 0.5, 100)   # 100 part-time, mean 3 juta
gaji_fulltime = np.random.normal(10, 1.5, 100)  # 100 full-time, mean 10 juta
gaji_semua = np.concatenate([gaji_parttime, gaji_fulltime])

print('=== Distribusi Bimodal: Part-time + Full-time ===')
print(f'  Mean   : Rp {np.mean(gaji_semua):.2f} juta')
print(f'  Median : Rp {np.median(gaji_semua):.2f} juta')
print()
print('⚠️ Mean dan Median keduanya ≈ 6.5 juta')
print('   Tapi TIDAK ADA karyawan yang gajinya sekitar 6.5 juta!')
print('   Part-time: ~3 juta, Full-time: ~10 juta')
print()
print('💡 Lesson: Jika distribusi bimodal, JANGAN pakai mean/median saja.')
print('   Harus analisis per-kelompok (stratified analysis)!')

# Visualisasi
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(gaji_semua, bins=30, color='#607D8B', edgecolor='black', alpha=0.7, density=True)
ax.axvline(np.mean(gaji_semua), color='red', linestyle='--', lw=2.5, label=f'Mean = {np.mean(gaji_semua):.1f} juta')
ax.axvline(np.median(gaji_semua), color='green', linestyle='-', lw=2.5, label=f'Median = {np.median(gaji_semua):.1f} juta')
ax.set_title('Distribusi Bimodal: Mean & Median Keduanya Menipu!', fontweight='bold', fontsize=13)
ax.set_xlabel('Gaji (juta Rp/bulan)')
ax.set_ylabel('Density')
ax.legend(fontsize=11)
ax.annotate('Part-time\n~3 juta', xy=(3, 0.15), fontsize=11, ha='center', color='blue', fontweight='bold')
ax.annotate('Full-time\n~10 juta', xy=(10, 0.1), fontsize=11, ha='center', color='blue', fontweight='bold')
ax.annotate('← Tidak ada yang\n   gajinya di sini!', xy=(6.5, 0.02), fontsize=10, ha='center', color='red')
plt.tight_layout()
plt.show()
````

### Key Takeaways:
1. **Mean** = paling informatif jika data simetris, tapi SENSITIF outlier

4. **Trimmed mean** = kompromi antara mean dan median

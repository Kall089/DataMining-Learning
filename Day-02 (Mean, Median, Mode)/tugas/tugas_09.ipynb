# Tugas 9: Analisis Data Bimodal

## Deskripsi
Data berikut adalah waktu tunggu pesanan di restoran cepat saji (menit).
Ada 2 jalur: drive-through (cepat) dan dine-in saat ramai (lambat).
Data tercampur TANPA label jalur.
Lakukan:
1. Hitung mean dan median dari data gabungan
2. Tunjukkan bahwa keduanya TIDAK mewakili waktu tunggu tipikal
3. Identifikasi 2 kelompok menggunakan threshold sederhana (misal 10 menit)
4. Hitung mean per kelompok
5. Jelaskan mengapa data bimodal perlu di-segmentasi

````python
import pandas as pd
import numpy as np

np.random.seed(42)

# Waktu tunggu (menit)
drive_through = np.array([3, 4, 5, 3, 4, 5, 6, 3, 4, 5, 4, 3, 5, 4, 6])  # cepat
dine_in_ramai = np.array([18, 20, 22, 19, 25, 21, 20, 23, 19, 22])          # lambat
waktu_semua = np.concatenate([drive_through, dine_in_ramai])
np.random.shuffle(waktu_semua)

print(f'Data waktu tunggu (n={len(waktu_semua)}): {waktu_semua}')
print(f'Min: {waktu_semua.min()}, Max: {waktu_semua.max()}')
````

```

Kelompok Cepat (< 10 menit): n=15, mean=4.27 menit

Kesimpulan:
Untuk data bimodal, JANGAN laporkan mean/median gabungan.
Laporkan per-segmen: "Drive-through ~4 menit, Dine-in ~21 menit"
```

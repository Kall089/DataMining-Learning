# Tugas 14: Analisis Segmentasi Data Campuran

## Deskripsi
Dataset berikut berisi durasi kunjungan website (menit) dari user mobile dan desktop.
Data TERCAMPUR tanpa label device.
Lakukan:
1. Hitung mean, median, mode dari data campuran
2. Tunjukkan bahwa mean/median MENIPU (jatuh di antara 2 cluster)
3. Segmentasi otomatis: cari threshold optimal menggunakan titik tengah 2 puncak histogram
4. Hitung statistik per segment
5. Buat visualisasi: histogram campuran + histogram per segment
6. Bandingkan: statistik gabungan vs statistik per-segment

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Mobile users: browsing singkat (1-5 menit)
mobile = np.random.normal(3, 0.8, 200).clip(0.5, 6)

# Desktop users: session panjang (12-20 menit)
desktop = np.random.normal(15, 2.5, 120).clip(8, 25)

# Data campuran (tanpa label)
durasi_semua = np.concatenate([mobile, desktop])
np.random.shuffle(durasi_semua)

print(f'Total data: {len(durasi_semua)} sessions')
print(f'Range: {durasi_semua.min():.1f} - {durasi_semua.max():.1f} menit')
````

````python
# TODO: Kerjakan di sini

# 1. Statistik gabungan

# 2. Tunjukkan misleading

# 3. Segmentasi (threshold = 8 menit sebagai titik tengah)

# 4. Statistik per segment

# 5. Visualisasi

# 6. Perbandingan
````

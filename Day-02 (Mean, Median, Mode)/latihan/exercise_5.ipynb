## 📋 Problem Statement

Kamu adalah data analyst di portal properti online. Tim editorial ingin membuat artikel: *"Berapa harga rumah di Jakarta sebenarnya?"*

Masalahnya:
- Harga rumah di Jakarta SANGAT bervariasi (dari Rp 500 juta di pinggiran sampai Rp 100 miliar di Menteng)
- Mean harga rumah pasti **di-inflate** oleh properti super mewah
- Median mungkin lebih tepat, tapi editorial ingin analisis per **wilayah**

Tugasmu:
1. Analisis harga properti per wilayah Jakarta
2. Tentukan ukuran central tendency yang paling tepat per wilayah
3. Buat rekomendasi harga "tipikal" yang bisa dipublikasikan
4. Identifikasi wilayah dengan distribusi harga paling skewed

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import trim_mean

np.random.seed(42)

# Generate data harga properti Jakarta (dalam miliar Rupiah)
def generate_harga(mean, std, n, outliers=[]):
    data = np.random.lognormal(np.log(mean), std, n)
    return np.concatenate([data, np.array(outliers)])

harga_jaksel = generate_harga(3.5, 0.6, 120, [25, 35, 50, 75])      # South Jakarta
harga_jakpus = generate_harga(5.0, 0.7, 80, [40, 60, 80, 100])      # Central Jakarta
harga_jakbar = generate_harga(2.0, 0.5, 100, [15, 20])              # West Jakarta
harga_jaktim = generate_harga(1.5, 0.4, 130, [10, 12])              # East Jakarta
harga_jakut = generate_harga(2.5, 0.6, 90, [18, 25, 30])            # North Jakarta

df_properti = pd.DataFrame({
    'harga_miliar': np.concatenate([harga_jaksel, harga_jakpus, harga_jakbar, harga_jaktim, harga_jakut]),
    'wilayah': (
        ['Jakarta Selatan'] * len(harga_jaksel) +
        ['Jakarta Pusat'] * len(harga_jakpus) +
        ['Jakarta Barat'] * len(harga_jakbar) +
        ['Jakarta Timur'] * len(harga_jaktim) +
        ['Jakarta Utara'] * len(harga_jakut)
    )
})

print(f'Total properti: {len(df_properti)}')
print(f'\nDistribusi per wilayah:')
print(df_properti['wilayah'].value_counts())
print(f'\nRange harga: Rp {df_properti["harga_miliar"].min():.2f}M - Rp {df_properti["harga_miliar"].max():.2f}M')
```

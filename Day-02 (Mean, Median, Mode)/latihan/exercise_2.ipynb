## 📋 Problem Statement

Kamu bekerja di divisi Quality Control sebuah toko online. Setiap hari, ratusan transaksi terjadi. Manajer ingin tahu apakah ada **anomali** dalam data penjualan harian — misalnya hari-hari di mana ada transaksi besar yang tidak biasa (possible fraud atau bulk purchase).

Salah satu cara mendeteksi anomali adalah melihat **gap antara mean dan median**:
- Jika gap besar → ada outlier yang menarik mean
- Jika gap kecil → data relatif "normal"

Tugasmu: Analisis data penjualan per hari dan identifikasi hari mana yang kemungkinan ada anomali.

## 🎯 Learning Goals
- [ ] Menghitung mean-median gap sebagai indikator outlier
- [ ] Menggunakan skewness untuk konfirmasi
- [ ] Memvisualisasi distribusi per kelompok
- [ ] Memberikan rekomendasi bisnis berdasarkan analisis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

# Generate data penjualan 7 hari
# Hari normal: transaksi Rp 50-200rb
# Hari anomali: ada beberapa transaksi > Rp 1 juta (bulk/fraud)

data_senin = np.random.normal(120, 30, 50).clip(50, 250)      # Normal
data_selasa = np.random.normal(110, 25, 55).clip(50, 200)     # Normal
data_rabu = np.concatenate([np.random.normal(100, 20, 45), [800, 1200, 1500]])  # ANOMALI!
data_kamis = np.random.normal(130, 35, 48).clip(50, 250)      # Normal
data_jumat = np.concatenate([np.random.normal(115, 28, 52), [950, 1100, 2000, 3000]])  # ANOMALI BESAR!
data_sabtu = np.random.normal(140, 30, 60).clip(60, 280)      # Normal
data_minggu = np.random.normal(125, 32, 45).clip(50, 260)     # Normal

# Buat dictionary
penjualan = {
    'Senin': data_senin,
    'Selasa': data_selasa,
    'Rabu': data_rabu,
    'Kamis': data_kamis,
    'Jumat': data_jumat,
    'Sabtu': data_sabtu,
    'Minggu': data_minggu,
}

print('Data Penjualan Toko Online (dalam ribu Rupiah):')
print('=' * 50)
for hari, data in penjualan.items():
    print(f'  {hari:7s}: {len(data)} transaksi, range [{data.min():.0f} - {data.max():.0f}]')
```

## 📋 Problem Statement

Kamu diminta oleh HR Director untuk menganalisis **distribusi gaji** di perusahaan teknologi "TechCorp". Perusahaan ini punya 4 departemen dengan struktur gaji yang berbeda:

- **Engineering**: Banyak level (junior → senior → lead → CTO)
- **Sales**: Gaji pokok rendah + komisi besar (beberapa sales top earner)
- **Operations**: Gaji relatif seragam
- **Executive**: Sedikit orang, gaji sangat tinggi

HR Director bertanya: *"Berapa gaji rata-rata perusahaan? Apakah karyawan kita dibayar cukup?"*

Kamu harus menunjukkan bahwa **"rata-rata" bisa menipu** dan memberikan angka yang lebih representatif.

## 🎯 Learning Goals
- [ ] Menganalisis distribusi gaji per departemen
- [ ] Mendemonstrasikan kapan mean menipu
- [ ] Memberikan rekomendasi ukuran yang tepat per kasus
- [ ] Membuat visualisasi yang menceritakan "cerita" data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import trim_mean

np.random.seed(42)

# Generate data gaji (dalam juta Rupiah/bulan)
gaji_engineering = np.concatenate([
    np.random.normal(8, 1.5, 30),   # Junior (30 orang)
    np.random.normal(15, 2, 20),    # Senior (20 orang)
    np.random.normal(25, 3, 8),     # Lead (8 orang)
    np.array([45, 55, 80])          # VP/CTO (3 orang)
]).clip(5, 100)

gaji_sales = np.concatenate([
    np.random.normal(6, 1, 35),     # Regular sales
    np.random.normal(12, 2, 10),    # Good performers
    np.array([35, 45, 60, 70, 90])  # Top salespeople (high commission)
]).clip(4, 100)

gaji_operations = np.random.normal(9, 1.5, 40).clip(6, 15)  # Seragam

gaji_executive = np.array([80, 95, 120, 150, 200])  # Eksekutif

# Gabung ke DataFrame
df_gaji = pd.DataFrame({
    'gaji': np.concatenate([gaji_engineering, gaji_sales, gaji_operations, gaji_executive]),
    'departemen': (
        ['Engineering'] * len(gaji_engineering) +
        ['Sales'] * len(gaji_sales) +
        ['Operations'] * len(gaji_operations) +
        ['Executive'] * len(gaji_executive)
    )
})

print(f'Total karyawan: {len(df_gaji)}')
print(f'Departemen: {df_gaji["departemen"].value_counts().to_dict()}')
print()
display(df_gaji.groupby('departemen')['gaji'].describe().round(2))
```

# ✅ Solution 1: Hitung & Bandingkan Mean vs Median

**Approach**: Hitung mean/median per kelas, analisis gap, visualisasi perbedaan  
**Key Insight**: Outlier menarik mean ke arahnya, sedangkan median tetap stabil di tengah.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

kelas_A = np.array([75, 78, 82, 80, 77, 81, 79, 83, 76, 79, 80, 82, 78, 81, 77])
kelas_B = np.array([75, 78, 82, 80, 77, 81, 79, 83, 76, 79, 80, 82, 78, 81, 10])
kelas_C = np.array([75, 78, 82, 80, 77, 81, 79, 83, 76, 79, 80, 82, 78, 81, 100])
```

## 📖 Penjelasan Pendekatan

1. Hitung mean dan median untuk setiap kelas
2. Gap = mean - median: indikator arah outlier
   - Gap positif → outlier di atas (menarik mean ke kanan)
   - Gap negatif → outlier di bawah (menarik mean ke kiri)
   - Gap ≈ 0 → distribusi relatif simetris
3. Pilih ukuran yang tepat berdasarkan ada/tidaknya outlier

```python
# SOLUSI TODO 1: Hitung mean dan median
df_stats = pd.DataFrame({
    'kelas': ['A', 'B', 'C'],
    'mean': [np.mean(kelas_A), np.mean(kelas_B), np.mean(kelas_C)],
    'median': [np.median(kelas_A), np.median(kelas_B), np.median(kelas_C)],
})
df_stats['gap'] = df_stats['mean'] - df_stats['median']
df_stats['gap_persen'] = (df_stats['gap'] / df_stats['median'] * 100).round(2)

print('Statistik per Kelas:')
display(df_stats)
print()
print('Interpretasi:')
print(f'  Kelas A: Gap = {df_stats.loc[0,"gap"]:.2f} → Distribusi SIMETRIS (tidak ada outlier)')
print(f'  Kelas B: Gap = {df_stats.loc[1,"gap"]:.2f} → Mean DITARIK KE BAWAH oleh outlier 10')
print(f'  Kelas C: Gap = {df_stats.loc[2,"gap"]:.2f} → Mean DITARIK KE ATAS oleh outlier 100')
```

```python
# SOLUSI TODO 2: Pilihan ukuran yang tepat
pilihan = {
    'A': ('mean', 'Distribusi simetris tanpa outlier, mean mewakili nilai tipikal'),
    'B': ('median', 'Ada outlier rendah (10) yang menarik mean turun, median lebih stabil'),
    'C': ('median', 'Ada outlier tinggi (100) yang menarik mean naik, median lebih representatif'),
}

print('Rekomendasi Ukuran Central Tendency:')
print('=' * 60)
for kelas, (ukuran, alasan) in pilihan.items():
    print(f'  Kelas {kelas}: Pilih {ukuran.upper()}')
    print(f'    Alasan: {alasan}')
    print()
```

```python
# SOLUSI TODO 3: Visualisasi
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

data_list = [kelas_A, kelas_B, kelas_C]
titles = ['Kelas A (Tanpa Outlier)', 'Kelas B (Outlier Rendah: 10)', 'Kelas C (Outlier Tinggi: 100)']
colors = ['#2196F3', '#FF9800', '#4CAF50']

for i, (data, title, color) in enumerate(zip(data_list, titles, colors)):
    axes[i].hist(data, bins=10, color=color, edgecolor='black', alpha=0.7)
    axes[i].axvline(np.mean(data), color='red', linestyle='--', lw=2, 
                    label=f'Mean = {np.mean(data):.1f}')
    axes[i].axvline(np.median(data), color='green', linestyle='-', lw=2,
                    label=f'Median = {np.median(data):.1f}')
    axes[i].set_title(title, fontweight='bold', fontsize=11)
    axes[i].set_xlabel('Nilai Ujian')
    axes[i].legend(fontsize=9)
    axes[i].set_xlim(0, 110)

axes[0].set_ylabel('Frekuensi')
plt.suptitle('Perbandingan Mean vs Median pada 3 Distribusi Berbeda', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 📌 Takeaways

- **Kelas A**: Mean ≈ Median → Kedua ukuran sama baiknya, mean sedikit lebih informatif
- **Kelas B**: Outlier rendah menarik mean ke bawah → Median lebih representatif
- **Kelas C**: Outlier tinggi menarik mean ke atas → Median lebih representatif
- **Rule of thumb**: Jika |gap| > 5% dari median, pertimbangkan pakai median
- Mean = median terjadi pada distribusi **sempurna simetris** (e.g., normal distribution tanpa outlier)

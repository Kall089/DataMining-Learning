# ✅ Solution 2: Membuat DataFrame dari Berbagai Sumber

**Approach**: pd.DataFrame() dari dict, list of dicts, pd.read_csv dengan StringIO  
**Key Insight**: Pandas menyediakan berbagai cara masuk — pilih sesuai format data sumbernya.

```python
import numpy as np
import pandas as pd
from io import StringIO

np.random.seed(42)

# Data sumber
data_pasien_lama = {
    'id_pasien': ['P001','P002','P003','P004','P005'],
    'nama':      ['Andi Wijaya','Budi Santoso','Citra Dewi','Dian Pratama','Eko Suryadi'],
    'umur':      [28, 45, 33, 52, 19],
    'berat_kg':  [65.5, 78.0, 55.2, 82.3, 60.0],
    'goldar':    ['A','O','B','AB','O'],
}
data_pemeriksaan = [
    {'id_pasien':'P001','tekanan_darah':'120/80','kolesterol':185,'diabetes':False},
    {'id_pasien':'P002','tekanan_darah':'140/90','kolesterol':240,'diabetes':True},
    {'id_pasien':'P003','tekanan_darah':'110/70','kolesterol':170,'diabetes':False},
    {'id_pasien':'P004','tekanan_darah':'160/100','kolesterol':280,'diabetes':True},
    {'id_pasien':'P005','tekanan_darah':'115/75','kolesterol':165,'diabetes':False},
]
csv_obat = """id_pasien,nama_obat,dosis_mg,frekuensi_hari
P002,Metformin,500,2
P004,Metformin,1000,2
P004,Amlodipine,5,1
P002,Atorvastatin,20,1"""

print('Sumber data siap!')
```

```python
# ============================================================
# SOLUSI TODO 1: Buat DataFrame dari setiap sumber
# ============================================================

# Metode 1: Dari dictionary
df_pasien = pd.DataFrame(data_pasien_lama)

# Metode 2: Dari list of dicts
df_periksa = pd.DataFrame(data_pemeriksaan)

# Metode 3: Dari CSV string dengan StringIO
df_obat = pd.read_csv(StringIO(csv_obat))

print('df_pasien:'); display(df_pasien)
print('df_periksa:'); display(df_periksa)
print('df_obat:'); display(df_obat)
```

```python
# ============================================================
# SOLUSI TODO 2: Tambahkan kolom BMI dan kategori_bmi
# ============================================================
tinggi_m = [1.72, 1.68, 1.60, 1.75, 1.65]

# Hitung BMI
df_pasien['tinggi_m'] = tinggi_m
df_pasien['bmi'] = (df_pasien['berat_kg'] / df_pasien['tinggi_m']**2).round(1)

# Kategorisasi BMI
def kategorikan_bmi(bmi):
    if bmi < 18.5:
        return 'Kurus'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obesitas'

df_pasien['kategori_bmi'] = df_pasien['bmi'].apply(kategorikan_bmi)

print('df_pasien dengan BMI dan kategori:')
display(df_pasien[['id_pasien','nama','berat_kg','tinggi_m','bmi','kategori_bmi']])
```

```python
# ============================================================
# SOLUSI TODO 3: Gabungkan df_pasien dan df_periksa
# ============================================================
df_gabung = pd.merge(df_pasien, df_periksa, on='id_pasien', how='inner')

print(f'df_gabung: {df_gabung.shape[0]} baris x {df_gabung.shape[1]} kolom')
display(df_gabung)

# Catatan: how='inner' → hanya baris yang ada di KEDUA dataframe
# how='left'  → semua baris kiri, NaN jika tidak ada di kanan
# how='outer' → semua baris dari kedua sisi
```

```python
# ============================================================
# SOLUSI TODO 4: Simpan dan baca kembali CSV
# ============================================================
df_gabung.to_csv('hasil_pasien.csv', index=False)
df_verify = pd.read_csv('hasil_pasien.csv')

if df_gabung.shape == df_verify.shape:
    print(f'Shape sama! {df_verify.shape} — CSV berhasil disimpan dan dibaca kembali')
else:
    print(f'BERBEDA! Asli: {df_gabung.shape}, Dibaca: {df_verify.shape}')

# Catatan: index=False mencegah kolom 'Unnamed: 0' muncul saat dibaca kembali
print()
print('Kenapa index=False?')
print('Jika index=True (default), pandas menyimpan index sebagai kolom.')
print('Saat dibaca kembali muncul kolom "Unnamed: 0" yang tidak perlu.')
```

## 📌 Takeaways

- `pd.DataFrame(dict)` → dari dictionary Python
- `pd.DataFrame(list_of_dicts)` → dari list berisi dict
- `pd.read_csv(StringIO(str))` → simulasi membaca CSV dari string
- `pd.merge(df1, df2, on='key', how='inner')` → gabung seperti JOIN di SQL
- Selalu gunakan `index=False` saat menyimpan CSV untuk menghindari kolom index tambahan
- `apply(fungsi)` → terapkan fungsi ke setiap baris/nilai dalam kolom
